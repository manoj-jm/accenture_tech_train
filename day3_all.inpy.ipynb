```python
import math

def toggle_bits(N):
    # Find the number of bits in the binary representation of N using log2
    num_bits = math.floor(math.log2(N)) + 1

    # Create a mask with all bits set to 1 up to the most significant bit (MSB)
    mask = (1 << num_bits) - 1

    # Toggle bits using XOR operator
    toggled_number = N ^ mask

    return toggled_number

# Test cases
print(toggle_bits(10))  # Output: 5
print(toggle_bits(12))  # Output: 3
```

```python
def find_equilibrium_point(arr):
    n = len(arr)
    total_sum = sum(arr)
    left_sum = 0    
    for i in range(n):
        if left_sum == total_sum - left_sum - arr[i]:
            return i
        left_sum += arr[i]
    return -1

# Test cases
print(find_equilibrium_point([-7, 1,5, 2, -4, 3, 0]))  # Output: 3
print(find_equilibrium_point([1, 2, 3]))  # Output: -1
```

```python
import math
def isprime(m):
    if m <= 1:
        return False
    if m == 2:
        return True
    if m % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(m)) + 1, 2):
        if m % i == 0:
            return False
    return True
##googly prime no 43 then yes ,4+3 -> 7 is prime 
n=43
isprime(n)
```
